```python
quality=df['quality']
values=[]
for k in quality:
  if(k<6.5):
    j="ordinary quality of wine"
    values.append(j)
  elif(k>6.5):
    j="High quality of wine"
    values.append(j)
values
```

```python
k=pd.DataFrame(values)
df['Wine_Quality']=k
df
```

```python
df2=pd.get_dummies(df)
df2
```

```python
x=df2.drop(['quality'],axis=1)
y=df2['quality']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators = 10,criterion = 'entropy',random_state=0)
```

```python
model.fit(x_train,y_train)
```

```python
pre=model.predict(x_test)
model.predict_proba(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
accuracy_score(y_test,pre)
```

```python
classification_report(y_test,pre)
```

```python
confusion_matrix(y_test,pre)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import decomposition, datasets
from sklearn import tree
from sklearn import ensemble
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV, cross_val_score
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
pca = decomposition.PCA()
randomforest= ensemble.RandomForestClassifier()
```

```python
pipe = Pipeline(steps =[('sc',sc),
                        ('pca',pca),
                        ('randomforest',randomforest)])
```

```python
n_components = list(range(1,x.shape[1]+1,1))
```

```python
criterion = ['gini', 'entropy']
max_depth = [2,4,6,8]
```

```python
parameters = dict(pca__n_components=n_components,
                 randomforest__criterion=criterion,
                 randomforest__max_depth=max_depth)
```

```python
clf = GridSearchCV(pipe, parameters)
```

```python
clf.fit(x,y)
```

```python
print('Best Criterion:',clf.best_estimator_.get_params()['randomforest__criterion'])
```

```python
print('Best max_depth:',clf.best_estimator_.get_params()['randomforest__max_depth'])
```

```python
print('Best No.of components:',clf.best_estimator_.get_params()['pca__n_components'])
```

```python
print(clf.best_estimator_.get_params()['randomforest'])
```

```python
CV_result=cross_val_score(clf,x,y,cv=4, n_jobs=-1)
```

```python
print(CV_result)
print(CV_result.mean())
print(CV_result.std())
```
